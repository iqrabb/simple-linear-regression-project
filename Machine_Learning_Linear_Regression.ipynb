```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python

# load dataset

df = pd.read_csv("/content/drive/MyDrive/student_exam_data_v1.csv")
df
```

```python
plt.scatter(df["study_hours"],df["exam_score"])
plt.xlabel("study_hours" , color = "red")
plt.ylabel("exam_score" , color = "red")
plt.title("student_exams_data" , color = "Green")
```

```python
# create regression model


LR = linear_model.LinearRegression()
```

```python

# train model

LR.fit(df[["study_hours"]] , df.exam_score)
```

```python
# predict exam scores

LR.predict([[5]])
```

```python
# model accuracy

LR.score(df[["study_hours"]] ,df.exam_score)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python

# regression line

plt.scatter(df["study_hours"],df["exam_score"] )
plt.xlabel("study_hours" , color = "blue")
plt.ylabel("exam_score" , color = "blue")
plt.title("student_exams_data" , color = "Green")

plt.plot(df.study_hours ,LR.predict(df[["study_hours"]]) , color = "red")

plt.show()
```
